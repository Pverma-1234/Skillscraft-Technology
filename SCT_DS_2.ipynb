```python
import pandas as pd

df = pd.read_csv('test.csv')

df.head()
```

```python
df_cleaned = df.drop(columns=['Cabin']).copy()

df_cleaned['Age'] = df_cleaned['Age'].fillna(df_cleaned['Age'].median())
df_cleaned['Fare'] = df_cleaned['Fare'].fillna(df_cleaned['Fare'].median())

df_cleaned['Sex'] = df_cleaned['Sex'].astype('category')
df_cleaned['Embarked'] = df_cleaned['Embarked'].astype('category')
df_cleaned.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(8, 5))
sns.histplot(df_cleaned['Age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()

plt.figure(figsize=(8, 5))
sns.histplot(df_cleaned['Fare'], bins=30, kde=True)
plt.title("Fare Distribution")
plt.show()

```

```python
# Age vs Pclass
plt.figure(figsize=(8, 5))
sns.boxplot(x='Pclass', y='Age', data=df_cleaned)
plt.title("Age Distribution Across Passenger Classes")
plt.show()


```

```python
# Fare vs Pclass
plt.figure(figsize=(8, 5))
sns.boxplot(x='Pclass', y='Fare', data=df_cleaned)
plt.title("Fare Distribution Across Passenger Classes")
plt.show()

```

```python
# Age vs Sex
plt.figure(figsize=(8, 5))
sns.boxplot(x='Sex', y='Age', data=df_cleaned)
plt.title("Age Distribution by Gender")
plt.show()
```

```python
# Numerical correlation
plt.figure(figsize=(8, 6))
sns.heatmap(df_cleaned.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```
