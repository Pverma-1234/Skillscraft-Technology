```python
import pandas as pd
df = pd.read_csv('cities_r2.csv')
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns




top10 = df.sort_values(by='population_total', ascending=False).head(10)


plt.figure(figsize=(14, 6))
bar_width = 0.4
x = range(len(top10))

plt.bar([i - bar_width/2 for i in x], top10['population_male'], width=bar_width, label='Male', color='steelblue')
plt.bar([i + bar_width/2 for i in x], top10['population_female'], width=bar_width, label='Female', color='lightcoral')

plt.xticks(ticks=x, labels=top10['name_of_city'], rotation=45)
plt.title('Male vs Female Population in Top 10 Indian Cities', fontsize=16, fontweight='bold')
plt.xlabel('City')
plt.ylabel('Population')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(12, 6))
sns.histplot(df['effective_literacy_rate_total'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Effective Literacy Rate Across Indian Cities', fontsize=16)
plt.xlabel('Effective Literacy Rate (%)', fontsize=12)
plt.ylabel('Number of Cities', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
